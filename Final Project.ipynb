```python
import pandas
```

```python
ndss = pandas.read_csv('../data/NDSS_data.csv')
ndss
```

```python
poa = pandas.read_csv('../data/POA list_Diabetes.csv')
poa
```

```python
ndss['PostCode']=pandas.to_numeric(ndss['PostCode'],errors='coerce')
ndss
```

```python

poa
```

```python

poa['HCP_count'] = poa.groupby('Post code')['MEDICAL PRACTITIONER / PHARMACY_ONE KEY ID'].transform(lambda x: x[x.str.contains('WAU')].count())
poa
```

```python
poa['Sum_call'] = poa.groupby('Post code')['PLANNED_CALLS_SEM1_2017'].transform(lambda x: x.sum())
poa
```

```python
df = pandas.merge(left =ndss, right=poa, how='inner',left_on='PostCode',right_on='Post code')
df
```

```python
df.corr()
```

```python
%matplotlib inline
df.plot.scatter (x='Type 2', y='HCP_count')
```
