```python
income = pandas.read_csv('../data/wage  by postcode.csv')
income
```

```python
extensive_df = pandas.merge(left=fullDf, right=income, how='inner',left_on='PostCode',right_on='Postcode')
extensive_df
```

```python
import sklearn.cluster
```

```python
clusterer = sklearn.cluster.KMeans(init='k-means++',n_clusters=3)
```

```python
interesting_columns=extensive_df[['Average salary and wages','Type 2']]
```

```python
Clusters = clusterer.fit_predict(interesting_columns)
Clusters
```

```python
import sklearn.metrics
```

```python
sklearn.metrics.silhouette_score(extensive_df[['Average salary and wages','Type 2']], Clusters)
```

```python
Colouring = numpy.array(['red','blue','green'])
x = extensive_df['Average salary and wages']
y = extensive_df['Type 2']
colour_by_cluster = Colouring[Clusters]
```

```python
matplotlib.pyplot.scatter(x,y, color=colour_by_cluster)
```

```python
scores = []
for i in range(2, 10):
    clustereri = sklearn.cluster.KMeans(n_clusters=i)
    cluster_suggestionsi = clustereri.fit_predict(interesting_columns)
    scores.append(sklearn.metrics.silhouette_score(interesting_columns, cluster_suggestionsi))
scores
```

```python
matplotlib.pyplot.plot(range(2, 10), scores)
```

```python
clusterer = sklearn.cluster.KMeans(init='k-means++',n_clusters=2)
```

```python
Clusters = clusterer.fit_predict(interesting_columns)
Colouring = numpy.array(['red','blue'])
x = extensive_df['Average salary and wages']
y = extensive_df['Type 2']
colour_by_cluster = Colouring[Clusters]
matplotlib.pyplot.scatter(x,y, color=colour_by_cluster)
```
