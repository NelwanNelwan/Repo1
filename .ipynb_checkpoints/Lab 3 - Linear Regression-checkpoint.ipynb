```python
# this allows plots to appear directly in the notebook
%matplotlib inline

# Import the libraries you'll need -- sklearn.linear_model, pandas and matplotlib.pyplot
import pandas
import sklearn.linear_model
import matplotlib.pyplot
```

```python
# Read in data from chicagohouseprices2.csv. Use column0 as an index
chicago_house = pandas.read_csv('chicagohouseprices2.csv',index_col=0)
chicago_house
```

```python
# Summarise the data with .describe()
chicago_house.describe()
```

```python
# Look for any linear correlations in the data - the bigger the number the better. Ordinary least squares will do
chicago_house.corr()
```

```python
chicago_house.Price.plot()
```

```python
# Make a scatter matrix. You might want to set the size of the figure really large so 
# that you can actually see the content
import math
import matplotlib.pyplot
ax=chicago_house.plot.scatter('Age','HouseSizeSqft', c=chicago_house.Price.map(math.log),
                        colormap=matplotlib.pyplot.cm.rainbow)
ax.set_xlabel( " Age ")
```

```python
# create a dataframe with just the estimated price column in it
chicago_house_estimated_price = chicago_house['EstimatedPrice']
chicago_house_estimated_price
```

```python
# create a RANSAC regressor
Ransac = sklearn.linear_model.RANSACRegressor()
```

```python
# use the RANSAC regressor to find how much of the Price is related to the EstimatedPrice
x = chicago_house[['Price']]
y = chicago_house.EstimatedPrice
Ransac.fit(x,y)
Ransac.estimator_.coef_
```

```python
# use that regression to predict Price based on EstimatedPrice
Predictions = Ransac.predict(x)
Predictions
```

```python
chicago_house.plot.scatter('Price','EstimatedPrice')
```

```python
# draw a scatter plot of Price / EstimatedPrice and overlay the RANSAC prediction on it
import matplotlib.pyplot
chicago_house.plot.scatter('Price','EstimatedPrice', c = Predictions,
                        colormap=matplotlib.pyplot.cm.rainbow)
```

```python
# RANSAC has an attribute called 'estimator'. What type is it?
Ransac.estimator_.intercept_
```

```python
# What is the coefficient (gradient) and y-intercept as reported by this estimator (hint: coef_ and intercept_)
# coef = array([ 0.97292556]) // intercept_ = 9773.619538148515
```

```python
# Create a new data frame with two columns: HouseSizeSqft and Bath
chicago_house2 = pandas.DataFrame({'HouseSizeSqft': chicago_house.HouseSizeSqft ,'Bath': chicago_house.Bath.fillna(value=0)})
chicago_house2 
```

```python
# Regress Price against these values
chicago_house.plot.scatter('HouseSizeSqft','Bath', c ='Price',
                        colormap=matplotlib.pyplot.cm.rainbow )
```

```python
Ransac_2 = sklearn.linear_model.RANSACRegressor()
x_2 = chicago_house2[['HouseSizeSqft','Bath']]
y_2 = chicago_house.Price
Ransac_2.fit(x_2,y_2)
Ransac_2.estimator_.coef_
```

```python
Ransac_2.estimator_.intercept_
```
