```python
import pandas_datareader.wb
```

```python
fi = pandas_datareader.wb.download(indicator='BX.KLT.DINV.CD.WD',country=['AU'],start=1970,end=2016)

fi.sort_index().plot()
```

```python
fi
```

```python
fi.columns = ['Foreign_Investment']
fi
```

```python
fi.reset_index (1, inplace= True)
fi.set_index('year', inplace=True)
fi
```

```python
fi.sort_index().plot()
```

```python
fi.describe()
```

```python
%matplotlib inline
lf = pandas_datareader.wb.download(indicator='SL.TLF.TOTL.IN',country=['AU'],start=1970,end=2016)

lf.sort_index().plot()
```

```python
lf.columns = ['Labour Force']
lf.reset_index (1, inplace= True)
lf.set_index('year', inplace=True)
lf
```

```python
import pandas
```

```python
au = pandas.concat([fi, lf],axis=1)
au
```

```python
ab = pandas_datareader.wb.download(indicator='BN.CAB.XOKA.CD',country=['AU'],start=1970,end=2016)
gdp = pandas_datareader.wb.download(indicator='NY.GDP.MKTP.CD',country=['AU'],start=1970,end=2016)
nm = pandas_datareader.wb.download(indicator='SM.POP.NETM',country=['AU'],start=1970,end=2016)
```

```python
ab.columns = ['Account_Balance']
ab.reset_index (1, inplace= True)
ab.set_index('year', inplace=True)

gdp.columns = ['GDP']
gdp.reset_index (1, inplace= True)
gdp.set_index('year', inplace=True)

nm.columns = ['Net_Migration']
nm.reset_index (1, inplace= True)
nm.set_index('year', inplace=True)

au = pandas.concat([fi,gdp,ab,lf,nm],axis=1)
au

```

```python
au.corr()
```

```python
au.describe()
```

```python
import sklearn.linear_model
Regressor = sklearn.linear_model.RANSACRegressor()
```

```python
X=au2[['GDP']]
y=au2.Foreign_Investment
```

```python
Regressor.fit(X,y)
```

```python
Predictions = Regressor.predict(X)
Predictions
```

```python
matplotlib.pyplot.scatter(X,y)
```

```python
matplotlib.pyplot.scatter(X,Predictions, color='red')
```

```python
Predictions = Regressor.predict(X)
Predictions
```
